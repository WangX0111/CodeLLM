将数据集拆分为训练集、验证集和测试集

```python
datasets_train_test = pass_datasets["train"].train_test_split(test_size=2)
datasets_train_validation = datasets_train_test["train"].train_test_split(test_size=2)

pass_datasets["train"] = datasets_train_validation["train"]
pass_datasets["validation"] = datasets_train_validation["test"]
pass_datasets["test"] = datasets_train_test["test"]

# DatasetDict({
#    train: Dataset({
#        features: ['title', 'text', 'url', 'authors', 'timestamp', 'tags'],
#        num_rows: 186368
#    })
#    validation: Dataset({
#        features: ['title', 'text', 'url', 'authors', 'timestamp', 'tags'],
#        num_rows: 3000
#    })
#    test: Dataset({
#        features: ['title', 'text', 'url', 'authors', 'timestamp', 'tags'],
#        num_rows: 3000
#    })
# })
```
